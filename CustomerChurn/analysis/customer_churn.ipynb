```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_excel(r"C:\Users\rsrmo\Documents\Data Analyst_ Portfolio\CustomerChurn\data\churn_data_cleaned.xlsx")
```

```python
df.describe()
```

```python
df.info()
```

```python
# Clean data
numeric_cols = ['CustomerID', 'Tenure Months', 'Monthly Charges', 'Total Charges']

# Convert only required to numeric
for col in numeric_cols:
    df[col] = pd.to_numeric(df[col], errors='coerce')


df['Churn Label'] = df['Churn Label'].map({'Yes':1, 'No':0})

# Churn rate
churn_rate = df['Churn Label'].mean() * 100
print(f"Overall Churn Rate: {churn_rate:.2f}%")
```

```python
df.info()
```

```python
# Churn by Contract
sns.barplot(x='Contract', y='Churn Label', data=df)
plt.title("Churn Rate by Contract Type")
plt.show()
```

```python
# Churn by Payment Method
sns.barplot(x='Payment Method', y='Churn Label', data=df)
plt.title("Churn Rate by Payment Method")
plt.show()
```

```python
# Tenure distribution
sns.histplot(data=df, x='Tenure Months', hue='Churn Label', kde=True)
plt.title("Tenure Distribution by Churn")
plt.show()
```
